```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### senttences
sentences=["i love my dog",
           "i love my cat",
           "you love my dog!",
           "do you think my dog is amazing"
           ] 
```

```python
## Define the vocabulary size
vocab_size=10000
```

```python
## one hot representation
one_hot_repr=[one_hot(words,vocab_size)for words in sentences]
one_hot_repr
   
```

```python
##word Embedding Representation

from tensorflow.keras.layers import Embedding 
from tensorflow.keras.utils import pad_sequences
from  tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sentence_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sentence_length)
embedded_docs
```

```python
## feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sentence_length))
# Build the model so layers (like Embedding) are created and parameter counts are known
model.build((None, sentence_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs[0:1])
```
